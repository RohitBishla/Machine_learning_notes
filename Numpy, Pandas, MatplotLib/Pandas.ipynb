```python
# reset our df to starting
df = iris.copy()
df.columns = ['sl', 'sw', 'pl', 'pw', 'flower_type']
df.describe()
```

```python
# adding new colume
df["diff_pl_pw"] = df["pl"] - df['pw']
df.head()
```

```python
df.head()
```

```python
# fill NaN we have two method 
# 1.  df.dropna(inplace = True) , It will del the NaN row
# 2.  Is to fill NaN with some value like mean, mode, or some random value.
```

```python
# Method 2
df.sw.fillna(df.sw.mean())
# not inplace to make inplace we have to inplace = True
# If we have to do mean of same flower type
a = df[df.flower_type == 'Iris-setosa']
df.sw.fillna(a.sw.mean(), inplace = True)
df.pl.fillna(a.pl.mean(), inplace = True)
print(df.head())
```

```python
dic = {}
for i in df.flower_type:
    if i in dic:
        dic[i]+=1
    else:
        dic[i]=1
print(dic)
```
