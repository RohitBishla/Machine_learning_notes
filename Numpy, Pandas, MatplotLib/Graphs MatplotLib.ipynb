```python
#  customising graphs by
# color, marker, linewidth, labels for axis/ dataset , title, axis, grid ,text
x = np.arange(0,5,0.1)
# np.arange(staring_point, ending_point, increment)
y = x**3
y2 = x**2
plt.plot(x,y2, color = 'red', linewidth = 5, label = "x^2")
# plt.plot(x,y,color = 'black', marker ='o')
# In this marker we will get marker as well as line
plt.plot(x,y,color = 'black', linewidth = 5, label = "x^3")

# to change axis plt.axis([starting_of_X, ending_of_x, starting_of_y, ending_of_y])
plt.axis([0,5, 0 ,50])

# We can also create gride just by calling function plt.gride()
plt.grid()

# to add text at random positions plt.text(x_position, y_position, "Any which you want to print")
plt.text(1,30, "Text", fontsize=20)
plt.ylabel("x^3")
plt.xlabel("x")
plt.title("Graph of x^3")
plt.legend()
plt.show()
# to show label of graph also we have to call a another function plt.legend() before plt.show()
```

```python
#  pie graph
size = [3,4,6,2]
labels =['A','B','C','D']
colors = ["blue","purple","red", "green"]
explode = [0.2, 0, 0 ,0]
plt.pie(size,colors=colors,labels = labels, explode = explode , autopct = '%0.2f%%', counterclock = False, startangle = 90)
plt.title("Pie graph")
plt.axis("equal")
plt.show()
```
